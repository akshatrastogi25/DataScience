```python
rating = google.iloc[:,1:]
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(rating)
data_scaled = pd.DataFrame(data_scaled, columns=rating.columns)
data_scaled.head()
```

```python
data = data_scaled.values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=17, color='r', linestyle='--')
```

```python
We look for the largest distance that we can vertically without crossing any horizontal line and this one is the red framed line on the above diagram. Let's count the number of lines cut by the red line drawn on the diagram and determine the optimal number of clusters. Cluster number will be 2 for this dataset.
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
rating2_segment = cluster.fit_predict(data)
```

```python
cluster.labels_
```

```python
rating2_segment
```

```python
plt.scatter(data[rating2_segment==0, 0], data[rating2_segment==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(data[rating2_segment==1, 0], data[rating2_segment==1, 1], s=100, c='blue', label ='Cluster 2')
plt.title('Clusters of USERS (Hierarchical Clustering Model)')
plt.xlabel('ratings')
plt.ylabel('based on categories')
plt.show()
```

```python
plt.scatter(data[:,0],data[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
google['Rating_segement'] = rating2_segment
```
