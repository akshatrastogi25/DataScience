# Q.2 What are the optimum number of principal components in PCA?


```python
from numpy import array
from sklearn.decomposition import PCA
```

```python
df.shape #we can see that there are 14 dimmensions
```

```python
# create the PCA instance
pca = PCA()
```

```python
# fit on data
pca.fit(df)
```

```python
# access values and vectors
print(pca.components_)
print(pca.explained_variance_)
```

```python
# transform data
B = pca.transform(df)
print(B)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

We can see that to achieve 100% variance we would need minimun 5 number of components.

# Q.3 Build a KNN classifier considering optimal number of principal components and value of K and state its score

#### Train Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, :13]
print(X)
y = df.iloc[:,13:]
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

#### Perform PCA on the train partition 

```python
pca = PCA(5) #WE SAW IN THE ABOVE GRAPH THAT WE CAN REDUCE OUR DATA TO 5 DIMMENSIONS WITH 100% VARIANCE
```

```python
# transform the train data with the PCA
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

#### Using KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

#### Predictions and Evaluations 

```python
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix, roc_auc_score, accuracy_score
```

```python
# Accuracy on test Data

print('Accuracy: ',metrics.accuracy_score(y_test, pred))
KNN = round(100*(metrics.accuracy_score(y_test, pred)),2)
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

# Q.4 Build a SVM Classifier and tune the hyperparameters to get the optimum model.

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

#### Predictions and Evaluations

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, classification_report
# Accuracy on test Data

print('Accuracy: ',metrics.accuracy_score(y_test, predictions))
SVM = round(100*(metrics.accuracy_score(y_test, predictions)),2)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

Woah! Notice that we are classifying everything into a single class! This means our model needs to have it parameters adjusted (it may also help to normalize the data).

Our model is only 64.81% accurate, and now we need to make it more accurate.

We can search for parameters using a GridSearch!

### Gridsearch 

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

You can inspect the best parameters found by GridSearchCV in the best_params_ attribute, and the best estimator in the best\_estimator_ attribute:

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, classification_report
# Accuracy on test Data

print('Accuracy: ',metrics.accuracy_score(y_test, grid_predictions))
SVM_GS = round(100*(metrics.accuracy_score(y_test, grid_predictions)),2)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

Using GridSearchCV we have increased our model accuracy from 64.81% to 83.33%.

Now our model can predict the customer segment with 83.33% accuracy rate.
