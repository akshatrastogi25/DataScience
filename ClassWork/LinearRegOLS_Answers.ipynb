```python
import numpy as np
X = np.array(df)
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print(vif)
X = pd.DataFrame(X)
```

```python
X = df.iloc[:,:61]
print(X)
```

```python
y = df.iloc[:,61:]
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 123) #give any name to random_state
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn import linear_model #Parametric Model
```

```python
regr = linear_model.LinearRegression()
```

```python
#Training
regr.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients', regr.coef_)
print('Intercept',regr.intercept_)
```

```python
#Testing
y_pred = regr.predict(X_test)
```

```python
y_pred
```

```python
#Mean Error Value
#Mean Squared Error 
#Evaluation
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test, y_pred)) #MSE
```

```python
print(r2_score(y_test, y_pred)) #Coefficient of Determination
```
