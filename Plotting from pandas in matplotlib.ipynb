```python
heart_dis=pd.read_csv("heart-disease.csv")
heart_dis.head()
```

```python
heart_dis["age"].plot.hist();
```

```python
heart_dis.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_45=heart_dis[heart_dis["age"]>45]
over_45
```

```python
len(over_45)
```

```python
#Pyplot
over_45.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
import matplotlib.pyplot as plt
```

```python
#oo method
fig,ax= plt.subplots(figsize=(10,6))
over_45.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```python
fig,ax= plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_45["age"],y=over_45["chol"],c=over_45["target"]);
ax.set(title="Heart Disease and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_45["chol"].mean(),linestyle="--");
```

```python
fig,(ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter = ax0.scatter(x=over_45["age"],y=over_45["chol"],c=over_45["target"],cmap="cool");
ax0.set(title="Heart Disease and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(over_45["chol"].mean(),linestyle="--");
scatter = ax1.scatter(x=over_45["age"],y=over_45["thalach"],c=over_45["target"]);
ax1.set(title="Heart Disease and Heart Rate Levels",xlabel="Age",ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(over_45["thalach"].mean(),linestyle="solid");
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
